**Introduction**

The present study aims to investigate the potential correlation between two variables: *alcohol consumption* and *income composition of resources*, focusing on Belgium, Germany and the Netherlands. The exploration could reveal insights into how economic factors might influence alcohol consumption patterns and how it differs between these neighbouring European countries.

*Research questions:*


1.   What's the correlation, if present, between income composition of resources and alcohol consumption?
2.   How does this correlation differ between Belgium, Germany and the Netherlands?



**Data Preparation**

```python
import pandas as pd
pd.set_option('display.max_columns', None)
HF=pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv")
HF.head()
```

```python
df = pd.DataFrame(HF)
```

```python
HF.columns
```

**Rows and columns relevant to our research question**

```python
Sel_HF= HF.loc[HF["Country"].isin(['Netherlands','Germany','Belgium']), ['Country','Year',
                                                                 'Income composition of resources','Alcohol']]
Sel_HF
```

**Explore and clean the data**

```python
#modify the DataFrame in place by renaming a variable
Sel_HF.rename(columns={'Income composition of resources': 'Income'}, inplace=True)
```

```python
df = pd.DataFrame(Sel_HF)
df
```

```python
min_year = Sel_HF['Year'].min()
max_year = Sel_HF['Year'].max()

# check if every year in between is included
years_included = set(Sel_HF['Year']) == set(range(Sel_HF['Year'].min(), Sel_HF['Year'].max() + 1))
print(f"Min year: {min_year}")
print(f"Max year: {max_year}")
print(f"Every year in between is included: {years_included}")
```

For each coutry, data on income composition of resources and alcohol consumption was included in the study from the year 2000 till 2015.

```python
Sel_HF.shape
```

```python
Sel_HF.isna().sum()  #count the number of missing values in the DataFrame
```

```python
Sel_HF[Sel_HF['Alcohol'].isna()]
```

For the year 2015 no data on alcohol consumption was found in the DataFrame for Belgium, Germany and the Netherlands, therefore values from the year 2015 were excluded from our data set.

```python
Sel_HF = Sel_HF.dropna() #remove the missing values
Sel_HF
```

**Descriptive statistics and visualisations**

```python
Sel_HF.drop('Year', axis=1).describe()
```

```python
Sel_HF.drop('Year', axis=1).hist(bins=5, figsize=(15,10))
```

The histogram for alcohol consumption revealed a possible outlier.

```python
Sel_HF_ALCOHOL= Sel_HF[Sel_HF['Alcohol']<8]
Sel_HF_ALCOHOL
```

After identifying the outlier we chose to remove it from the data set.

```python
outlier = Sel_HF[(Sel_HF['Country'] == 'Netherlands') & (Sel_HF['Year'] == 2014) & (Sel_HF['Alcohol'] == 0.01)].index

Sel_HF_cleaned = Sel_HF.drop(outlier)
```

```python
Sel_HF_cleaned
```

**Descriptive statistics for 'Income' and 'Alcohol' grouped by 'Country'**

```python
descriptive_stats = Sel_HF_cleaned.groupby('Country')[['Income', 'Alcohol']].describe()
descriptive_stats
```

**Belgium**: Has a relatively stable income composition with a moderate range. Alcohol consumption varies, with a mean around 11.

**Germany**: Shows higher variability in both income composition and alcohol consumption. The mean alcohol consumption is slightly higher than in Belgium.

**Netherlands**: Has a stable income composition with a lower mean. Alcohol consumption is lower than in Belgium and Germany, with less variability.

```python
import altair as alt
```

**Alcohol consumption per year in each country**

```python
alt.Chart(Sel_HF_cleaned).mark_circle().encode(
    x=alt.X('Year:O', title='Year'),  # Specify ordinal scale for the 'Year' variable
    y=alt.Y('Alcohol', scale=alt.Scale(domain=[8, 13])),
    color='Country',
)


```

Germany and the Netherlands show a generally decreasing trend in alcohol consumption per capita (15+) over the years. Belgium shows a fluctuating trend with a slight decrease over the years 2006-2012, followed by an increase.

```python
alt.Chart(Sel_HF).mark_circle().encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Income', scale=alt.Scale(domain=[0.8, 0.96])),
    color='Country'
)
```

All three countries show a generally increasing trend in income composition of resources over the years. Compared to the Netherlands and Germany, Belgium reported a lower income from the year 2006.

```python
box_plot = alt.Chart(Sel_HF_cleaned).mark_boxplot().encode(
    x='Country',
    y=alt.Y('Income:Q', title='Income', scale=alt.Scale(domain=[0.8, 0.96])),
    color='Country'
).properties(width=400)  # Adjust the width

box_plot
```

Germany shows higher variability in income composition compared to Netherlands and Belgium where it is more stable.

**Interactive scatter plot corralating 'Income composition of resources' and 'Alcohol'**

```python
interactive_scatter = alt.Chart(Sel_HF_cleaned).mark_circle().encode(
    x=alt.X('Income', scale=alt.Scale(domain=[0.85, 0.95])),
    y=alt.Y('Alcohol', scale=alt.Scale(domain=[8, 13])),
    color='Country',
    tooltip=['Year', 'Income', 'Alcohol', 'Country']
).properties(width=600, height=400).interactive()

interactive_scatter
```

For Germany, there is an evident negative correlation between alcohol consumption and income, suggesting that as income increases, alcohol consumption tends to decrease. A similar trend is noted for the Netherlands, whereas for Belgium no correlation is found.

**Conclusion**

Our analysis reveals distinct patterns in income composition and alcohol consumption across Belgium, Germany, and the Netherlands.

A negative correlation was found between income composition and alcohol consumption, particularly evident in Germany and the Netherlands.
Belgium shows a more stable income with varied alcohol consumption trends, which does not allow for a clear correlation between the two.
